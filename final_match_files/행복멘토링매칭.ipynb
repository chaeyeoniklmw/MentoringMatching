```python
import pandas as pd

# 시간대 문자열을 파싱하여 요일과 시간을 반환하는 함수
def parse_time_slots(time_slot_str):
    days = {'월': 0, '화': 1, '수': 2, '목': 3, '금': 4, '토': 5, '일': 6}
    parsed_slots = []

    for time_slot in time_slot_str.replace('/', ',').split(','):
        time_slot = time_slot.strip()

        if len(time_slot) < 3:
            continue

        day = time_slot[0]
        time_range = time_slot[2:]

        if day in days:
            start_end = time_range.split('-')
            if len(start_end) == 2:
                try:
                    start = int(start_end[0].strip())
                    end = int(start_end[1].strip())
                    for hour in range(start, end + 1):
                        parsed_slots.append((days[day], hour))
                except ValueError:
                    continue
            else:
                try:
                    hour = int(time_range.strip())
                    parsed_slots.append((days[day], hour))
                except ValueError:
                    continue

    return set(parsed_slots)

# 멘토와 멘티 매칭
matches = []
matched_mentees = set()
matched_mentors = set()

for _, mentor in mentors.iterrows():
    mentor_time_slots = parse_time_slots(mentor['멘토링 시간대'])
    mentee_count = 0
    mentee_names = []  # 멘티 이름을 저장할 리스트

    for _, mentee in mentees.iterrows():
        if mentee['이름'] in matched_mentees:
            continue

        mentee_time_slots = parse_time_slots(mentee['멘토링 시간대'])

        if mentor_time_slots & mentee_time_slots:  # 교집합이 있으면 매칭
            mentee_names.append(mentee['이름'])  # 멘티 이름 추가
            matched_mentees.add(mentee['이름'])
            matched_mentors.add(mentor['이름'])
            mentee_count += 1

            # 한 멘토에 두 명의 멘티만 매칭
            if mentee_count >= 2:
                break

    if mentee_names:  # 매칭된 멘티가 있는 경우
        matches.append({
            '멘토 이름': mentor['이름'],
            '멘티 이름': ', '.join(mentee_names),  # 멘티 이름을 한 줄로 결합
            '멘토 전공': mentor['나의 전공'],
            '멘티 전공': ', '.join([mentees[mentees['이름'] == name]['나의 전공'].values[0] for name in mentee_names])  # 멘티 전공도 결합
        })

# 매칭 결과 데이터프레임 생성
matching_results = pd.DataFrame(matches)

# 매칭 안 된 멘토 데이터프레임 생성
unmatched_mentors = mentors[~mentors['이름'].isin(matched_mentors)]
unmatched_mentors['멘티 이름'] = '매칭 안 됨'  # 매칭 안 된 멘토에 대한 멘티 이름 설정
unmatched_mentors['멘토 전공'] = unmatched_mentors['나의 전공']
unmatched_mentors['멘티 전공'] = '매칭 안 됨'  # 매칭 안 된 멘티 전공 설정

# 매칭 결과와 매칭되지 않은 멘토 데이터프레임 병합
final_results = pd.concat([matching_results, unmatched_mentors[['이름', '멘티 이름', '멘토 전공', '멘티 전공']]], ignore_index=True)

# 매칭 결과 출력
print(final_results)

# 총 매칭 팀 수
num_teams = len(matching_results)
print(f"총 매칭 팀 수: {num_teams}")

# 결과를 엑셀 파일로 저장
final_results.to_excel('시간대매칭_결과.xlsx', index=False)
```

```python
# 전공 유사성 정의
def is_similar_major(mentor_major, mentee_major):
    similar_fields = {
        '생명환경': ['생명환경', '식품영양학과', '식품공학과', '생명환경공학'],
        '패션산업학과': ['패션산업학과', '현대미술', '아동학과/예술심리치료', '원예생명조경', '도시환경예술디자인'],
        '경제학과': ['경제학과', '사회과학대학(열린전공)', '경영학과', '미래산업융합대학 내 유사학과'],
        '언론영상학부': ['언론영상학부 비즈니스커뮤니케이션 전공', '언론영상학부-디지털영상전공', '비즈니스커뮤니케이션전공'],
        '데이터사이언스': ['데이터사이언스', '소프트웨어융합학과', '정보보호학부'],
        '교육심리학과': ['교육심리학과'],
        '기독교학과': ['기독교학과'],
        '공예': ['공예']
    }

    # 멘토의 전공에 해당하는 유사 전공 목록을 가져오기
    for field, majors in similar_fields.items():
        if mentor_major in majors:
            return mentee_major in majors
    return False

# 멘토와 멘티 매칭 수정
matches = []
matched_mentees = set()  # 중복 방지를 위한 멘티 집합
matched_mentors = set()  # 매칭된 멘토 집합

for _, mentor in mentors.iterrows():
    mentor_time_slots = parse_time_slots(mentor['멘토링 시간대'])

    for _, mentee in mentees.iterrows():
        if mentee['이름'] in matched_mentees:
            continue

        mentee_time_slots = parse_time_slots(mentee['멘토링 시간대'])

        # 멘토 전공과 멘티 희망 전공이 일치하거나 유사 전공일 때
        if (mentor['나의 전공'] == mentee['희망전공'] or
            is_similar_major(mentor['나의 전공'], mentee['희망전공'])) and (mentor_time_slots & mentee_time_slots):
            matches.append({
                '멘토 이름': mentor['이름'],
                '멘티 이름': mentee['이름'],
                '멘토 전공': mentor['나의 전공'],
                '멘티 전공': mentee['나의 전공'],
                '매칭 시간대': mentor_time_slots & mentee_time_slots
            })
            matched_mentees.add(mentee['이름'])
            matched_mentors.add(mentor['이름'])

            # 한 멘토에 두 명의 멘티만 매칭
            if len(matched_mentees) % 2 == 0:
                break

# 매칭 결과 출력 및 저장
matching_results = pd.DataFrame(matches)
print(matching_results)

# 매칭된 팀 수
num_teams = len(matching_results) // 2
print(f"총 매칭 팀 수: {num_teams}")

# 매칭 안 된 멘토 및 멘티 출력
unmatched_mentors = mentors[~mentors['이름'].isin(matched_mentors)]
unmatched_mentees = mentees[~mentees['이름'].isin(matched_mentees)]

print("\n매칭 안 된 멘토:")
print(unmatched_mentors)

print("\n매칭 안 된 멘티:")
print(unmatched_mentees)

# 결과를 엑셀 파일로 저장
matching_results.to_excel('멘토링_매칭_결과1.xlsx', index=False)
```
