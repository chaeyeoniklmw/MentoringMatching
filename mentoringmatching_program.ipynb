#데이터 생성
- 이름,시간대,흥미,사는지역,취미,친목,전공,MBTI반영

```python
import pandas as pd
import random

# 랜덤 데이터 생성 함수
def generate_random_time():
    times = [
        "월요일 오전 9시-11시", "월요일 오후 2시-4시",
        "화요일 오전 10시-12시", "화요일 오후 1시-3시",
        "수요일 오전 9시-11시", "수요일 오후 2시-4시",
        "목요일 오전 10시-12시", "목요일 오후 1시-3시",
        "금요일 오전 9시-11시", "금요일 오후 2시-4시",
        "토요일 오전 10시-12시", "토요일 오후 1시-3시",
        "일요일 오전 11시-1시", "일요일 오후 2시-4시"
    ]
    return random.choice(times)

def generate_random_interest():
    interests = [
        "동아리", "유학", "해외연수", "패션디자인",
        "스포츠", "음악", "미술", "프로그래밍",
        "뷰티", "영상 편집"
    ]
    return random.choice(interests)

def generate_random_location():
    locations = [
        "노원구", "강남구", "마포구", "서초구",
        "강동구", "성북구", "동작구", "송파구",
        "중랑구", "광진구"
    ]
    return random.choice(locations)

def generate_random_hobby():
    hobbies = [
        "독서", "영화 감상", "요리", "등산",
        "게임", "여행", "피아노 연주", "사진 촬영",
        "수영", "축구", "자전거 타기"
    ]
    return random.choice(hobbies)

def generate_random_social():
    socials = [
        "친구 만나기", "카페 탐방", "저녁 모임",
        "네트워킹", "워크숍 참여", "팀 활동",
        "온라인 모임", "동아리 활동", "집들이", "파티 참여",
        "연합동아리", "사진동아리"
    ]
    return random.choice(socials)

def generate_random_name(index):
    # 성과 이름의 조합을 늘려서 다양화
    first_names = ["김", "이", "박", "최", "정", "강", "조", "윤", "임", "오", "장", "한", "류", "백", "원"]
    last_names = ["민수", "지훈", "서연", "하늘", "지우", "현우", "지호", "유진", "예은", "태훈", "나래", "성민", "지원", "소연", "도현"]
    return f"{random.choice(first_names)}{random.choice(last_names)}_{index}"

# 멘토 30명, 멘티 60명 생성
mentors_mentees = ['멘토'] * 30 + ['멘티'] * 60

# 희망전공 목록
majors = [
    '소프트웨어융합학과', '경영학', '문헌정보학과', '바이오헬스융합학과', '아동학과',
    '언론영상학과', '교육심리학과', '영어영문학과', '독일문화콘텐츠학과', '디지털미디어학과',
    '원예생명조경학과', '국어국문학과', '산업디자인학과', '패션산업학과',
    '식품공학과', '심리학', '데이터사이언스학과'
]

# 데이터 샘플 생성
data = {
    '이름': [generate_random_name(i) for i in range(90)],  # 이름 열 생성
    '멘토/멘티': random.sample(mentors_mentees, len(mentors_mentees)),
    '멘토링 시간대': [generate_random_time() for _ in range(90)],
    '희망전공': [None] * 90,  # 초기값 설정
    '나의 전공': [None] * 90,  # 초기값 설정
    '관심분야': [generate_random_interest() for _ in range(90)],
    '취미': [generate_random_hobby() for _ in range(90)],
    '친목': [generate_random_social() for _ in range(90)],
    '사는 지역': [generate_random_location() for _ in range(90)],
    'MBTI유형': [random.choice(['INTJ', 'ENFP', 'ENTJ', 'INFP', 'ESTJ', 'ESFP', 'ISFJ', 'ISTP']) for _ in range(90)]
}

# 멘토와 멘티의 희망전공 및 나의 전공 설정
for i in range(90):
    if data['멘토/멘티'][i] == '멘티':
        data['희망전공'][i] = random.choice(majors)
    else:
        data['희망전공'][i] = None
    data['나의 전공'][i] = random.choice(majors)  # 모든 사람에게 나의 전공 할당

# 데이터프레임 생성
df = pd.DataFrame(data)

# '이름' 열을 첫 번째 컬럼으로 이동
df = df[['이름', '멘토/멘티', '멘토링 시간대', '희망전공', '나의 전공', '관심분야', '취미', '친목', '사는 지역', 'MBTI유형']]

# Excel 파일로 저장
df.to_excel('mentoring_data_sample.xlsx', index=False)

# 결과 확인
print(df.head(10))  # 처음 10개의 데이터만 출력하여 확인
```

# K-means 군집화를 통해 팀매칭에 필요한 요소를 기반으로 군집화 진행


### 1) '멘토링 시간대', '희망전공', '관심분야', '취미', '친목', '사는 지역', 'MBTI유형' 반영 팀매칭

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans

# 데이터 불러오기
df = pd.read_excel('mentoring_data_sample.xlsx')

# 필요한 열만 선택
columns_of_interest = ['멘토/멘티', '멘토링 시간대', '희망전공', '관심분야', '취미', '친목', '사는 지역', 'MBTI유형']
data = df[columns_of_interest]

# 문자열 데이터를 숫자 데이터로 변환
label_encoders = {}
for column in data.columns[1:]:  # '멘토/멘티'를 제외한 열들
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column].astype(str))
    label_encoders[column] = le

# 스케일링
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data.drop('멘토/멘티', axis=1))

# K-평균 군집화
k = 10  # 군집의 수는 데이터와 목적에 따라 조정
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(scaled_data)

# 데이터프레임에 군집 정보 추가
df['클러스터'] = clusters

# 멘토와 멘티 분리
mentors = df[df['멘토/멘티'] == '멘토']
mentees = df[df['멘토/멘티'] == '멘티']

# 군집별로 멘토와 멘티를 매칭
matches = []
remaining_mentees = []
remaining_mentors = []

for cluster in range(k):
    cluster_mentors = mentors[mentors['클러스터'] == cluster]
    cluster_mentees = mentees[mentees['클러스터'] == cluster]

    while not cluster_mentors.empty and len(cluster_mentees) >= 2:
        # 멘토와 2명의 멘티를 매칭
        mentor = cluster_mentors.iloc[0]
        mentee1 = cluster_mentees.iloc[0]
        mentee2 = cluster_mentees.iloc[1]

        matches.append({
            'Mentor': mentor['이름'],
            'Mentee1': mentee1['이름'],
            'Mentee2': mentee2['이름'],
            'Cluster': cluster
        })

        # 사용한 멘토와 멘티는 목록에서 제거
        cluster_mentors = cluster_mentors.iloc[1:]
        cluster_mentees = cluster_mentees.iloc[2:]

    if not cluster_mentees.empty:
        remaining_mentees.extend(cluster_mentees.to_dict('records'))

    if not cluster_mentors.empty:
        remaining_mentors.extend(cluster_mentors.to_dict('records'))

# 남은 멘토와 멘티를 랜덤 매칭
remaining_mentees_df = pd.DataFrame(remaining_mentees)
remaining_mentors_df = pd.DataFrame(remaining_mentors)

# 랜덤으로 매칭하기
np.random.shuffle(remaining_mentors_df.values)
remaining_mentees_df = remaining_mentees_df.reset_index(drop=True)
remaining_mentors_df = remaining_mentors_df.reset_index(drop=True)

min_len = min(len(remaining_mentees_df) // 2, len(remaining_mentors_df))

for i in range(min_len):
    mentor = remaining_mentors_df.loc[i, '이름']
    mentee1 = remaining_mentees_df.loc[2 * i, '이름']
    mentee2 = remaining_mentees_df.loc[2 * i + 1, '이름'] if 2 * i + 1 < len(remaining_mentees_df) else None
    matches.append({
        'Mentor': mentor,
        'Mentee1': mentee1,
        'Mentee2': mentee2,
        'Cluster': '랜덤'
    })

# 매칭 결과를 데이터프레임으로 변환
matching_df = pd.DataFrame(matches)

# Excel 파일로 저장
matching_df.to_excel('mentoring_matching_results.xlsx', index=False)

# 결과 확인
print(matching_df.head(10))  # 처음 10개의 데이터만 출력하여 확인
```

### 2) 시간대 별 팀매칭

```python
import pandas as pd
import numpy as np

# 데이터 불러오기
df = pd.read_excel('mentoring_data_sample.xlsx')

# 멘토와 멘티 분리
mentors = df[df['멘토/멘티'] == '멘토']
mentees = df[df['멘토/멘티'] == '멘티']

# 시간대별로 멘토와 멘티를 매칭
matches = []
remaining_mentees = []
remaining_mentors = []

# 시간대별로 그룹화
mentor_groups = mentors.groupby('멘토링 시간대')
mentee_groups = mentees.groupby('멘토링 시간대')

# 시간대가 같은 그룹끼리 매칭
for time_slot, mentor_group in mentor_groups:
    if time_slot in mentee_groups.groups:
        mentee_group = mentee_groups.get_group(time_slot)

        while not mentor_group.empty and not mentee_group.empty:
            mentor = mentor_group.iloc[0]
            mentee1 = mentee_group.iloc[0]
            mentee2 = mentee_group.iloc[1] if len(mentee_group) > 1 else None

            matches.append({
                'Mentor': mentor['이름'],
                'Mentee1': mentee1['이름'],
                'Mentee2': mentee2['이름'] if mentee2 is not None else None,
                'Time Slot': time_slot
            })

            # 사용한 멘토와 멘티는 목록에서 제거
            mentor_group = mentor_group.iloc[1:]
            mentee_group = mentee_group.iloc[2:] if mentee2 is not None else mentee_group.iloc[1:]

        if not mentee_group.empty:
            remaining_mentees.extend(mentee_group.to_dict('records'))

        if not mentor_group.empty:
            remaining_mentors.extend(mentor_group.to_dict('records'))

# 남은 멘토와 멘티를 랜덤 매칭
remaining_mentees_df = pd.DataFrame(remaining_mentees)
remaining_mentors_df = pd.DataFrame(remaining_mentors)

# 랜덤으로 매칭하기
np.random.shuffle(remaining_mentors_df.values)
remaining_mentees_df = remaining_mentees_df.reset_index(drop=True)
remaining_mentors_df = remaining_mentors_df.reset_index(drop=True)

min_len = min(len(remaining_mentees_df) // 2, len(remaining_mentors_df))

for i in range(min_len):
    mentor = remaining_mentors_df.loc[i, '이름']
    mentee1 = remaining_mentees_df.loc[2 * i, '이름']
    mentee2 = remaining_mentees_df.loc[2 * i + 1, '이름'] if 2 * i + 1 < len(remaining_mentees_df) else None
    matches.append({
        'Mentor': mentor,
        'Mentee1': mentee1,
        'Mentee2': mentee2,
        'Time Slot': '랜덤'
    })

# 매칭 결과를 데이터프레임으로 변환
matching_df = pd.DataFrame(matches)

# Excel 파일로 저장
matching_df.to_excel('mentoring_matching_results.xlsx', index=False)

# 결과 확인
print(matching_df.head(10))  # 처음 10개의 데이터만 출력하여 확인
```
